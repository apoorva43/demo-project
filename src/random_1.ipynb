```python
def max_three(a, b, c):
    return max(max(a, b), c)
```

```python
max_three(10, -20, 30)
```

```python
def add_list(nums):
    if len(nums) == 0:
        return 0
    return nums[0] + add_list(nums[1:])
```

```python
add_list([1,2,3,4,5])
```

```python
def mul_list(nums):
    if len(nums) == 0:
        return 1
    return nums[0] * mul_list(nums[1:])
```

```python
mul_list([1,2,3,4,5])
```
